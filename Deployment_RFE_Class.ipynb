```python
import pickle
import joblib
loaded_model = pickle.load(open("finalized_model_svmlinear.sav",'rb'))

#joblib.dump(classifier, "finalized_model_linear.pkl")
#loaded_model = joblib.load("model.pkl")

scaled_model = joblib.load("scaler.pkl")
```

```python
## RF has selected these columns ['hrmo', 'sg_c', 'dm_yes']
hrmo = float(input("Please enter the hrmo value: "))
sg_c = int(input("If sg type is 'c', then enter 1 else enter 0:\n"))
dm_yes = int(input("If dm is present, then enter 1 else enter 0:\n"))
```

```python
preinput = scaled_model.transform([[hrmo, sg_c, dm_yes]])
result = loaded_model.predict(preinput)
result
```
